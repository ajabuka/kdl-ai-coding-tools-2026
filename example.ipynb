# Text Analysis Toolkit — Example Usage

This notebook demonstrates how to use the text analysis pipeline to analyse word frequencies in early modern English texts.

## Overview

The pipeline performs the following steps:
1. Loads a text file from the `data/` directory
2. Removes punctuation and stop words (including early modern English stop words)
3. Builds bigram and trigram models to capture multi-word phrases
4. Lemmatises tokens using spaCy
5. Computes word frequencies
6. Generates an interactive HTML histogram of the top N words

## Expected Inputs

- **Text file**: A plain text file in the `data/` directory (e.g. `Hamlet.txt` or `test_data.txt`)
- **Stop word file**: `data/earlyModernStopword.txt` (loaded automatically)
- **Configurable variables**: Set in the main analysis cell (details below)

## Expected Outputs

- An interactive HTML histogram saved to the `results/` directory
- The chart displays the top N words, their counts, and percentage contributions
- The chart also renders inline in the notebook

---

To run the full analysis, open `text_analysis_notebook.ipynb` and execute all cells.

## Configuration Reference

The main analysis cell in `text_analysis_notebook.ipynb` exposes these configurable variables:

| Variable | Default | Description |
|---|---|---|
| `n` | `10` | Number of top words to display |
| `documentName` | `'Hamlet.txt'` | Input text file in `data/` directory |
| `outputFile` | `'topTenPlainText'` | Base name for the output HTML file |
| `stop_language` | `'english'` | Language for nltk stop words |
| `figureTitle` | `'Top 10 Words, Hamlet'` | Title of the histogram |

To analyse a different text:

1. Place your `.txt` file in the `data/` directory
2. Change `documentName` to your file's name
3. Update `figureTitle` to match your text
4. Re-run all cells

## Testing with the Provided Test Data

A small test file `data/test_data.txt` is included for quick verification:

1. Open `text_analysis_notebook.ipynb`
2. Set `documentName = 'test_data.txt'` and `figureTitle = 'Top 10 Words, Test Data'`
3. Run all cells
